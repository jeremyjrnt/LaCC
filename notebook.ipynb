What are the Accuracy scores?

Define a function that computes rank-based accuracy score, then, iterate over the 18 folds. For each fold,  train the decoder **using the `learn_decoder` function** (the function is already imported from `learn_decoder.py`) on the fold train data, obtain the predicions on the fold test data, and store both the accuracy score of each concept (use the labels from `concepts`) as well as the average score of the 10 concepts.  

```
# TODO

import numpy as np

def cosine_similarity(x: np.ndarray, y: np.ndarray) -> float:
    norm_x = np.linalg.norm(x)
    norm_y = np.linalg.norm(y)
    if norm_x == 0 or norm_y == 0:
        return 0.0  # or np.nan if you prefer
    return np.dot(x, y) / (norm_x * norm_y)


def rank_score(pred_vec: np.ndarray,
               true_idx: int,
               all_semantic_vecs: np.ndarray) -> int:
    """
    Computes the rank of the true semantic vector among all concepts,
    based on cosine similarity to the predicted vector.
    """
    similarities = [
        cosine_similarity(pred_vec, vec) for vec in all_semantic_vecs
    ]
    ordered = np.argsort(similarities)[::-1]  # descending order
    rank = np.where(ordered == true_idx)[0][0] + 1  # 1-based rank
    return rank
```

```
def crossval_decode(X: np.ndarray,
                    Y: np.ndarray,
                    concept_labels: np.ndarray,
                    n_folds: int = 18):
    """
    X : (180, V)  imaging data
    Y : (180, 300) semantic vectors
    concept_labels : (180,) list of strings
    """
    n_concepts = X.shape[0]
    fold_size  = n_concepts // n_folds

    # Store per-concept ranks and per-fold averages
    concept_ranks = np.full(n_concepts, np.nan)
    fold_avg      = []

    indices = np.arange(n_concepts)

    for fold in range(n_folds):
        # ----------------------------------------------------------
        # Split indices
        # ----------------------------------------------------------
        test_mask  = slice(fold*fold_size, (fold+1)*fold_size)
        test_idx   = indices[test_mask]
        train_idx  = np.setdiff1d(indices, test_idx, assume_unique=True)

        # ----------------------------------------------------------
        # Train decoder on 170 concepts
        # ----------------------------------------------------------
        M = learn_decoder(X[train_idx], Y[train_idx])   # outputs (V, 300)

        # ----------------------------------------------------------
        # Decode the 10 test concepts
        # ----------------------------------------------------------
        preds = X[test_idx] @ M                         # (10, 300)

        # ----------------------------------------------------------
        # Compute rank for each test concept
        # ----------------------------------------------------------
        ranks_this_fold = []
        for local_i, global_i in enumerate(test_idx):
            r = rank_score(preds[local_i], global_i, Y)
            concept_ranks[global_i] = r
            ranks_this_fold.append(r)

        # ----------------------------------------------------------
        # Fold-level average rank
        # ----------------------------------------------------------
        fold_avg.append(np.mean(ranks_this_fold))

        print(f"Fold {fold+1:02d}  |  mean rank = {fold_avg[-1]:5.2f}")

    # --------------------------------------------------------------
    # Turn outputs into convenient Python objects
    # --------------------------------------------------------------
    fold_scores = np.array(fold_avg)

    # map concept → rank
    concept_rank_dict = {label: int(rank)
                         for label, rank in zip(concept_labels,
                                                concept_ranks)}

    return fold_scores, concept_rank_dict

fold_scores, per_concept = crossval_decode(data, vectors, concepts)

# 1.  All 18 fold averages
print("\nFold-level average ranks:")
for i, s in enumerate(fold_scores, 1):
    print(f"  Fold {i:02d}: {s:5.2f}")

# 2.  Example: rank of a specific concept
print("\nRank of concept 'dog':", per_concept['dog'])

# 3.  Quick sanity check – global mean should be close to the
#     mean of fold averages.
print("\nGlobal mean rank:", np.nanmean(list(per_concept.values())))
```

```
import pickle
import numpy as np
import pandas as pd

def load_exp_data(pkl_path, ranks):
    # Load the experiment .pkl file
    with open(pkl_path, 'rb') as f:
        data = pickle.load(f)

    # Extract and flatten topic IDs (one per passage)
    topic_ids = np.array(data['labelsPassageCategory']).flatten()

    # Extract passage ID per sentence
    passage_ids = np.array(data['labelsPassageForEachSentence']).flatten()

    # Normalize topic IDs if they start at 1 instead of 0
    if topic_ids.min() == 1:
        topic_ids = topic_ids - 1

    # Flatten the topic name list: each is a single-element array
    key_categories = [str(item[0]) for item in data['keyPassageCategory'][0]]

    # Handle invalid passage IDs (should be ≤ max index of topic_ids)
    max_valid_pid = len(topic_ids) - 1
    valid_indices = [i for i, pid in enumerate(passage_ids) if pid <= max_valid_pid]

    # Keep only valid entries
    passage_ids = [passage_ids[i] for i in valid_indices]
    ranks = [ranks[i] for i in valid_indices]

    # Map each passage ID to its topic ID
    passage_to_topic = {pid: topic_ids[pid] for pid in range(len(topic_ids))}
    sentence_topic_ids = [passage_to_topic[pid] for pid in passage_ids]

    # Build DataFrame
    df = pd.DataFrame({
        'sentence_index': np.arange(len(ranks)),
        'rank': ranks,
        'topic_id': sentence_topic_ids
    })

    # Compute mean rank per topic
    topic_df = df.groupby('topic_id')['rank'].mean().reset_index()
    topic_df['topic_name'] = topic_df['topic_id'].apply(lambda tid: key_categories[tid])

    return topic_df.sort_values(by='rank')


ranks_exp2 = np.load("ranks_384.npy")  # or your computed rank list for EXP2
exp2_topic_ranks = load_exp_data("EXP2.pkl", ranks_exp2)

print("EXP2 Topic-Level Average Ranks:")
print(exp2_topic_ranks)

ranks_exp3 = np.load("ranks_243.npy")  # or your computed rank list for EXP3
exp3_topic_ranks = load_exp_data("EXP3.pkl", ranks_exp3)

print("EXP3 Topic-Level Average Ranks:")
print(exp3_topic_ranks)
```

```
# Download and extract the data and learn_decoder.py
# !wget --load-cookies /tmp/cookies.txt "https://docs.google.com/uc?export=download&confirm=$(wget --quiet --save-cookies /tmp/cookies.txt --keep-session-cookies --no-check-certificate 'https://docs.google.com/uc?export=download&id=1xZaorRH-xxjfochvSesAhOTUg82_Xq56' -O- | sed -rn 's/.*confirm=([0-9A-Za-z_]+).*/\1\n/p')&id=1xZaorRH-xxjfochvSesAhOTUg82_Xq56" -O files.zip && rm -rf /tmp/cookies.txt
!wget --header="Host: drive.usercontent.google.com" --header="Accept: text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7" --header="Accept-Language: en-US,en;q=0.9,he;q=0.8" --header="Cookie: HSID=AHJfxja1o67aaDDKP; SSID=AcFaYUEeiC88MwrF9; APISID=-FXvHmBvJ828Jrpq/AaIp_RI6gKwBAA-zy; SAPISID=_psqReiv0O2WdiVv/AhLpZThJtVNAPgAJP; __Secure-1PAPISID=_psqReiv0O2WdiVv/AhLpZThJtVNAPgAJP; __Secure-3PAPISID=_psqReiv0O2WdiVv/AhLpZThJtVNAPgAJP; S=billing-ui-v3=pX9aAWC8SzxQZfQvQ-0SbCFRz65PPkVY:billing-ui-v3-efe=pX9aAWC8SzxQZfQvQ-0SbCFRz65PPkVY:maestro=dsv3G-owxPD6uTATLH0lBQZNadhFo6ZKJiuB9usoQVU; __Secure-BUCKET=CPgG; SID=g.a000kggtmVDh8D92rqHe5fiG-bMoXQw7Ld8Tf_C8qHhSE2ZoFUyx_uObP_F4bCqI8I561ccGMwACgYKAWESARQSFQHGX2Mi5DnhBiJ2gjjbMSP0XJbU5BoVAUF8yKrlBjWMdNOfGnmA7TZzmbWD0076; __Secure-1PSID=g.a000kggtmVDh8D92rqHe5fiG-bMoXQw7Ld8Tf_C8qHhSE2ZoFUyx2BFINS8lXhFUyAFwuvl8CQACgYKAW4SARQSFQHGX2MiWd6bHkI0JN89-1dFZUbS2hoVAUF8yKpc-H3AD8N6tj-dmFG21SeE0076; __Secure-3PSID=g.a000kggtmVDh8D92rqHe5fiG-bMoXQw7Ld8Tf_C8qHhSE2ZoFUyxJl_TGsCsjeiVN72q3lSCWQACgYKASASARQSFQHGX2MiULluXa7aABDwxgCWjB6IyhoVAUF8yKoy_HHYLCqIFMwNjx-GwYWe0076; __Secure-ENID=20.SE=jyM_w2hA8DW6FvPOh9wudde93a0A9P41Epzo098LV_LyU79-VVcJ9K-vNLrhCLuVzi69CyV4RxlSls8AAT9J8odwIXi_ISVn8Z1U1DH52BC3YiwOwO9LKUsBesCbGx2D6u1XwZ5GIP_PZMo1tkLLJq2VCtcxRP9OtC_QgHNbAD4eyc1TTu1C8XbZLFTOIgb0k9IfM2bMBXeha6t3sJysARZWpDIzs3I8wWZ5JtABB253grtjQyCnxyy9MUgTcYAVaoEGwgVHV4V4lSY6gydFkO2gYxl7JqYloqCq74HahGK54TBlsGZIOTM_KvFAsIidcrPaVOBpH6IGQTPChxy3Tr-GLK7VpBiQ8JW7V0xC8XTN1crEaaZnGFQ6MrjDv8f3hCY0Kg; AEC=AQTF6HwEtUB747fVHMzvOWJV9pmRoGs8Ix8FJ1HTrxbE9NY1dtyro2AvNQ; NID=515=Wdt0NWZqVSh3TtdIfjXCGTCCkj7jaJjt-lkOL3hLD_hPSSMyGxKkVthECwGGFbbxmvfM2iKZ1SkPGDVgLwjghAOrV9Ya7iEJJ0eSXZSfszc0WxRXm3Jy6LxqPEZLmY8v3AIkMX-o8KE5ZRXGEzgv_s9pfgS8bmeiIGT13Iiyw9tPzRZDChGijNbZ0Mp1oF-4YKikOZCyo8Km9wXOgLAC9dbeIqAlTdER97cQ7B5GajyRLH_bFrg0lCVN4tyZEycjihHOu6Eq_V88rswgV7uvzemJ_yk4WbbIWJVm9NCO4tWdDQG8NY3EY57xAJbmIhu260jDftYwzjCnpqJ8C1iCm-FjboF6xJwKJEtLkCXagIcSWxfPGqRWIn5KY72ogAMZTlUZ5RE5F8bH4sFgkt5pW_AalY5mxYPOfZgF-9hcJYsF71rMOic6mqSfvR8iNo-k6_SZ-4o5WkYxbwdLgiaIOiCJHkhIGBoXsm5hh5BHDqlk5ERGnFn5zpqOguNLJFjXT3nhaP1g_a2fFvd0bmZw2A9Y6tBNAC7CbDOmSHSmYmLag0qVcqu286CZh5svuhdM-QPcSCt5u0kPgfWN3KBha0G9L9qCiDIwntvnlVNoUYLBM4je1bhGjO9M2tdH_vteLo4vjm9Cq-4I2A; __Secure-1PSIDTS=sidts-CjIB3EgAEi6AJoaJlu_IOdqmparuSFUne3RqD5YKK5hcqKjRlc0CTp9lSpyH2OoVVoqmlxAA; __Secure-3PSIDTS=sidts-CjIB3EgAEi6AJoaJlu_IOdqmparuSFUne3RqD5YKK5hcqKjRlc0CTp9lSpyH2OoVVoqmlxAA; SIDCC=AKEyXzW7IJ8miV8hX_pqzqqPW7--eMWuWfausspLBoDPlfZDCRZDED56ohpancLYOBPizfUzi1aM; __Secure-1PSIDCC=AKEyXzVGnmSuG07J22njRVFPQ_sk88MgnqtYxJd-M0_9Pz3jdh4GpGPhPOqCMrayTU9SJTW3n54; __Secure-3PSIDCC=AKEyXzVScJYbKdtIImPYKpTRkExsc5UhC5n9Rkk8wNFlMZNW3_xkvWlimAXWaZ4T7kTcJy5AE4I" --header="Connection: keep-alive" "https://drive.usercontent.google.com/download?id=1xZaorRH-xxjfochvSesAhOTUg82_Xq56&export=download&authuser=0&confirm=t&uuid=efeb9ce5-a5c5-453b-938d-6c0ece963f3c&at=APZUnTV18b5mSao0MQ2JbtpefTxr%3A1719665236172" -c -O 'files.zip'
!unzip files.zip
!rm files.zip

#Let's load the functions from learn_decoder.py
from learn_decoder import *

concepts = np.genfromtxt('stimuli_180concepts.txt', dtype=np.dtype('U')) #The names of the 180 concepts
vectors = read_matrix("vectors_180concepts.GV42B300.txt", sep=" ")

import numpy as np
!pip install gensim
from gensim.models import KeyedVectors

import gensim.downloader as api


def cosine_similarity(x: np.ndarray, y: np.ndarray) -> float:
    norm_x = np.linalg.norm(x)
    norm_y = np.linalg.norm(y)
    if norm_x == 0 or norm_y == 0:
        return 0.0  # or np.nan if you prefer
    return np.dot(x, y) / (norm_x * norm_y)


def rank_score(pred_vec: np.ndarray,
               true_idx: int,
               all_semantic_vecs: np.ndarray) -> int:
    """
    Computes the rank of the true semantic vector among all concepts,
    based on cosine similarity to the predicted vector.
    """
    similarities = [
        cosine_similarity(pred_vec, vec) for vec in all_semantic_vecs
    ]
    ordered = np.argsort(similarities)[::-1]  # descending order
    rank = np.where(ordered == true_idx)[0][0] + 1  # 1-based rank
    return rank

def crossval_decode(X: np.ndarray,
                    Y: np.ndarray,
                    concept_labels: np.ndarray,
                    n_folds: int = 18):
    """
    X : (180, V)  imaging data
    Y : (180, 300) semantic vectors
    concept_labels : (180,) list of strings
    """
    n_concepts = X.shape[0]
    fold_size  = n_concepts // n_folds

    # Store per-concept ranks and per-fold averages
    concept_ranks = np.full(n_concepts, np.nan)
    fold_avg      = []

    indices = np.arange(n_concepts)

    for fold in range(n_folds):
        # ----------------------------------------------------------
        # Split indices
        # ----------------------------------------------------------
        test_mask  = slice(fold*fold_size, (fold+1)*fold_size)
        test_idx   = indices[test_mask]
        train_idx  = np.setdiff1d(indices, test_idx, assume_unique=True)

        # ----------------------------------------------------------
        # Train decoder on 170 concepts
        # ----------------------------------------------------------
        M = learn_decoder(X[train_idx], Y[train_idx])   # outputs (V, 300)

        # ----------------------------------------------------------
        # Decode the 10 test concepts
        # ----------------------------------------------------------
        preds = X[test_idx] @ M                         # (10, 300)

        # ----------------------------------------------------------
        # Compute rank for each test concept
        # ----------------------------------------------------------
        ranks_this_fold = []
        for local_i, global_i in enumerate(test_idx):
            r = rank_score(preds[local_i], global_i, Y)
            concept_ranks[global_i] = r
            ranks_this_fold.append(r)

        # ----------------------------------------------------------
        # Fold-level average rank
        # ----------------------------------------------------------
        fold_avg.append(np.mean(ranks_this_fold))

        print(f"Fold {fold+1:02d}  |  mean rank = {fold_avg[-1]:5.2f}")

    # --------------------------------------------------------------
    # Turn outputs into convenient Python objects
    # --------------------------------------------------------------
    fold_scores = np.array(fold_avg)

    # map concept → rank
    concept_rank_dict = {label: int(rank)
                         for label, rank in zip(concept_labels,
                                                concept_ranks)}

    return fold_scores, concept_rank_dict, M
```

```
import numpy as np
import pandas as pd
import glob
import os
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
from learn_decoder import *

# -------------------------------------------------
# Helpers
# -------------------------------------------------
def rank_score(pred, gold_index, all_targets):
    sims = cosine_similarity(pred.reshape(1, -1), all_targets).flatten()
    rank = (-sims).argsort().tolist().index(gold_index) + 1
    return rank

def crossval_17_1_decode(X_test, Y, decoder, n_folds=18, test_transform=None):
    """
    Evaluate a fixed decoder with 18 folds (10 concepts per fold) on the test participant.
    If test_transform is provided, apply it to X_test before using the decoder.
    """
    if test_transform is not None:
        X_eval = test_transform(X_test)
    else:
        X_eval = X_test

    n = X_eval.shape[0]
    fold_size = n // n_folds
    indices = np.arange(n)
    fold_ranks = []

    # Note: training is done outside; this function only evaluates a fixed decoder.
    for fold in range(n_folds):
        test_mask = slice(fold * fold_size, (fold + 1) * fold_size)
        test_idx = indices[test_mask]
        preds = X_eval[test_idx] @ decoder
        for k, idx in enumerate(test_idx):
            fold_ranks.append(rank_score(preds[k], idx, Y))

    return np.mean(fold_ranks)

# -------------------------------------------------
# Aggregation utilities
# -------------------------------------------------
def compute_positive_shift(train_matrices, eps=1e-6):
    """
    Per-feature shift δ so that all training values become strictly positive:
      δ_j = max(0, -min_s,c X[s][c,j]) + eps
    train_matrices: list of (C x V) arrays for the 14 training subjects.
    Returns δ with shape (V,).
    """
    stacked = np.stack(train_matrices, axis=0)  # (S, C, V)
    min_per_feature = stacked.min(axis=(0,1))   # (V,)
    delta = np.maximum(0.0, -min_per_feature) + eps
    return delta

def aggregate_arithmetic(train_matrices):
    """
    Element-wise arithmetic mean across subjects: (C x V)
    """
    return np.mean(np.stack(train_matrices, axis=0), axis=0)

def aggregate_geometric_shifted(train_matrices, delta):
    """
    Element-wise geometric mean on shifted data:
      GM_j = exp( mean_s log( X_s[:,j] + δ_j ) )
    Returns a (C x V) array in the 'shifted' space (no un-shift).
    """
    shifted = [X + delta for X in train_matrices]       # ensure strictly positive
    log_vals = [np.log(Xs) for Xs in shifted]
    mean_log = np.mean(np.stack(log_vals, axis=0), axis=0)
    return np.exp(mean_log)

def aggregate_harmonic_shifted(train_matrices, delta):
    """
    Element-wise harmonic mean on shifted data:
      HM_j = S / (sum_s 1/(X_s[:,j] + δ_j))
    Returns a (C x V) array in the 'shifted' space (no un-shift).
    """
    shifted = [X + delta for X in train_matrices]
    inv_vals = [1.0 / Xs for Xs in shifted]
    mean_inv = np.mean(np.stack(inv_vals, axis=0), axis=0)
    return 1.0 / mean_inv

# -------------------------------------------------
# Load semantic vectors and concept labels
# -------------------------------------------------
vectors = np.loadtxt("vectors_180concepts.GV42B300.txt")  # (180 x 300)
concepts = np.genfromtxt("stimuli_180concepts.txt", dtype=str)

# -------------------------------------------------
# Load participant data
# -------------------------------------------------
file_paths = sorted(glob.glob("M*_avg_pca.csv"))
participant_ids = [os.path.basename(f).split('_')[0] for f in file_paths]
participant_data = {pid: pd.read_csv(f).values for pid, f in zip(participant_ids, file_paths)}
n_participants = len(participant_ids)

# -------------------------------------------------
# LOSO over test participants; evaluate 3 aggregations
# -------------------------------------------------
methods = ["arithmetic", "geometric", "harmonic"]
results = {m: [] for m in methods}

for test_pid in participant_ids:
    train_ids = [pid for pid in participant_ids if pid != test_pid]
    X_train_list = [participant_data[pid] for pid in train_ids]  # 14 subjects, each (180 x V)
    X_test_raw = participant_data[test_pid]                      # (180 x V)

    # ===== Arithmetic mean =====
    X_avg_arith = aggregate_arithmetic(X_train_list)  # (180 x V)
    dec_arith = learn_decoder(X_avg_arith, vectors)

    mr_arith = crossval_17_1_decode(
        X_test=X_test_raw,
        Y=vectors,
        decoder=dec_arith,
        n_folds=18,
        test_transform=None  # no shift for arithmetic mean
    )
    results["arithmetic"].append({"test_participant": test_pid, "mean_rank": round(mr_arith, 2)})

    # ===== Geometric mean (shifted) =====
    delta_geo = compute_positive_shift(X_train_list, eps=1e-6)   # (V,)
    X_avg_geo = aggregate_geometric_shifted(X_train_list, delta_geo)  # (180 x V)
    dec_geo = learn_decoder(X_avg_geo, vectors)

    # test transform: add the SAME delta used for training aggregation
    def test_transform_geo(X):
        return X + delta_geo

    mr_geo = crossval_17_1_decode(
        X_test=X_test_raw,
        Y=vectors,
        decoder=dec_geo,
        n_folds=18,
        test_transform=test_transform_geo
    )
    results["geometric"].append({"test_participant": test_pid, "mean_rank": round(mr_geo, 2)})

    # ===== Harmonic mean (shifted) =====
    delta_harm = compute_positive_shift(X_train_list, eps=1e-6)  # (V,)
    X_avg_harm = aggregate_harmonic_shifted(X_train_list, delta_harm)  # (180 x V)
    dec_harm = learn_decoder(X_avg_harm, vectors)

    def test_transform_harm(X):
        return X + delta_harm

    mr_harm = crossval_17_1_decode(
        X_test=X_test_raw,
        Y=vectors,
        decoder=dec_harm,
        n_folds=18,
        test_transform=test_transform_harm
    )
    results["harmonic"].append({"test_participant": test_pid, "mean_rank": round(mr_harm, 2)})

    print(f"✓ {test_pid}: mean rank — arithmetic={mr_arith:.2f}, geometric={mr_geo:.2f}, harmonic={mr_harm:.2f}")

# -------------------------------------------------
# Results tables
# -------------------------------------------------
df_arith = pd.DataFrame(results["arithmetic"])
df_geo   = pd.DataFrame(results["geometric"])
df_harm  = pd.DataFrame(results["harmonic"])

# Merge for plotting
merged = df_arith.merge(df_geo, on="test_participant", suffixes=("_arith", "_geo"))
merged = merged.merge(df_harm, on="test_participant")
merged = merged.rename(columns={"mean_rank": "mean_rank_harm"})

# Compute global means
gmean_arith = merged["mean_rank_arith"].mean()
gmean_geo   = merged["mean_rank_geo"].mean()
gmean_harm  = merged["mean_rank_harm"].mean()

print("\n🌐 Global mean rank (lower = better):")
print(f"  Arithmetic: {gmean_arith:.2f}")
print(f"  Geometric : {gmean_geo:.2f}")
print(f"  Harmonic  : {gmean_harm:.2f}")

# -------------------------------------------------
# Plot: grouped bars per held-out participant
# -------------------------------------------------
plt.figure(figsize=(14, 6))
x = np.arange(len(merged["test_participant"]))
w = 0.25

plt.bar(x - w, merged["mean_rank_arith"], width=w, label=f"Arithmetic (μ={gmean_arith:.2f})")
plt.bar(x,      merged["mean_rank_geo"],   width=w, label=f"Geometric (μ={gmean_geo:.2f})")
plt.bar(x + w,  merged["mean_rank_harm"],  width=w, label=f"Harmonic (μ={gmean_harm:.2f})")

plt.axhline(90, linestyle="--", color="red", label="Chance (90)")
plt.ylabel("Mean Rank (Lower = Better)")
plt.xlabel("Held-out Participant")
plt.title("LOSO (14/1) with 17/1 evaluation • Aggregation comparison")
plt.xticks(x, merged["test_participant"], rotation=45)
plt.legend()
plt.tight_layout()
plt.savefig("aggregation_comparison_LOSO_meanrank.png")
plt.show()

# Save CSV
merged.to_csv("aggregation_comparison_LOSO_meanrank.csv", index=False)
print("✅ Saved: aggregation_comparison_LOSO_meanrank.csv and aggregation_comparison_LOSO_meanrank.png")
```

```
import numpy as np
import pandas as pd
import glob
import os
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import TwoSlopeNorm
from learn_decoder import *

# -----------------------------
# Helper functions
# -----------------------------
def rank_score(pred, gold_index, all_targets):
    sims = cosine_similarity(pred.reshape(1, -1), all_targets).flatten()
    rank = (-sims).argsort().tolist().index(gold_index) + 1
    return rank

def crossval_17_1_decoder_eval(X_train, Y_train, X_test, Y_test, n_folds=18):
    """
    18 folds = 10 concepts per fold (180 concepts total).
    Train decoder on the 17 folds (170 concepts), predict on held-out fold,
    return mean rank (lower = better).
    """
    n_concepts = X_train.shape[0]
    fold_size = n_concepts // n_folds
    indices = np.arange(n_concepts)
    fold_ranks = []

    for fold in range(n_folds):
        test_mask = slice(fold * fold_size, (fold + 1) * fold_size)
        test_idx = indices[test_mask]
        train_idx = np.setdiff1d(indices, test_idx, assume_unique=True)

        M = learn_decoder(X_train[train_idx], Y_train[train_idx])
        preds = X_test[test_idx] @ M

        for i, idx in enumerate(test_idx):
            fold_ranks.append(rank_score(preds[i], idx, Y_test))

    return np.mean(fold_ranks)

def add_mean_excl_diag(df):
    """Add MEAN row/col excluding the diagonal."""
    # Row means (train-wise)
    row_means = [df.iloc[i, :].drop(df.columns[i]).mean() for i in range(df.shape[0])]
    # Column means (test-wise)
    col_means = [df.iloc[:, j].drop(df.index[j]).mean() for j in range(df.shape[1])]

    out = df.copy()
    out["MEAN"] = row_means
    out.loc["MEAN"] = col_means + [np.nan]
    return out

def offdiag_values(mat: np.ndarray) -> np.ndarray:
    """Return flattened off-diagonal values from a square matrix."""
    n = mat.shape[0]
    mask = ~np.eye(n, dtype=bool)
    return mat[mask]

# -----------------------------
# Load semantic vectors and concept labels
# -----------------------------
vectors = np.loadtxt("vectors_180concepts.GV42B300.txt")
concepts = np.genfromtxt("stimuli_180concepts.txt", dtype=str)

# -----------------------------
# Load PCA-reduced data from participants
# -----------------------------
file_paths = sorted(glob.glob("M*_avg_pca.csv"))
participant_ids = [os.path.basename(f).split('_')[0] for f in file_paths]
participant_data = [pd.read_csv(f).values for f in file_paths]

# -----------------------------
# TRUE transfer matrix
# -----------------------------
n = len(participant_ids)
score_matrix_true = np.zeros((n, n))

for i in range(n):  # train participant
    for j in range(n):  # test participant
        score = crossval_17_1_decoder_eval(
            X_train=participant_data[i],
            Y_train=vectors,
            X_test=participant_data[j],
            Y_test=vectors
        )
        score_matrix_true[i, j] = score
        print(f"[TRUE] Decoder {participant_ids[i]} → Test {participant_ids[j]}: mean rank = {score:.2f}")

score_df_true = pd.DataFrame(score_matrix_true, index=participant_ids, columns=participant_ids)
score_df_true_with_avg = add_mean_excl_diag(score_df_true)

# -----------------------------
# PERMUTED-LABELS transfer matrix (single permutation)
# -----------------------------
rng = np.random.default_rng(12345)  # reproducible
perm_idx = rng.permutation(vectors.shape[0])
perm_vectors = vectors[perm_idx]

score_matrix_perm = np.zeros((n, n))
for i in range(n):  # train participant
    for j in range(n):  # test participant
        score = crossval_17_1_decoder_eval(
            X_train=participant_data[i],
            Y_train=perm_vectors,  # permuted labels for training
            X_test=participant_data[j],
            Y_test=vectors         # true labels for evaluation
        )
        score_matrix_perm[i, j] = score
        print(f"[PERM1] Decoder {participant_ids[i]} → Test {participant_ids[j]}: mean rank = {score:.2f}")

score_df_perm = pd.DataFrame(score_matrix_perm, index=participant_ids, columns=participant_ids)
score_df_perm_with_avg = add_mean_excl_diag(score_df_perm)

# -----------------------------
# Plot: TRUE heatmap (define color scale)
# -----------------------------
plt.figure(figsize=(11, 9))
valid_vals_true = score_df_true_with_avg.values[~np.isnan(score_df_true_with_avg.values)]
norm_true = TwoSlopeNorm(vmin=np.min(valid_vals_true), vcenter=90, vmax=np.max(valid_vals_true))

sns.heatmap(score_df_true_with_avg, annot=True, fmt=".1f", cmap="coolwarm",
            norm=norm_true, mask=score_df_true_with_avg.isna(),
            cbar_kws={'label': 'Mean Rank'})
plt.title("Cross-Participant 17/1 Transfer (Mean Rank)\nTRUE labels • MEAN excludes diagonal")
plt.xlabel("Test Participant")
plt.ylabel("Train Participant")
plt.tight_layout()
plt.savefig("decoder_transfer_matrix_TRUE.png", dpi=200)
plt.show()

# -----------------------------
# Plot: PERM heatmap (use SAME color scale as TRUE)
# -----------------------------
plt.figure(figsize=(11, 9))
sns.heatmap(score_df_perm_with_avg, annot=True, fmt=".1f", cmap="coolwarm",
            norm=norm_true, mask=score_df_perm_with_avg.isna(),
            cbar_kws={'label': 'Mean Rank'})
plt.title("Cross-Participant 17/1 Transfer (Mean Rank)\nPermuted labels (1 perm) • SAME scale as TRUE • MEAN excludes diagonal")
plt.xlabel("Test Participant")
plt.ylabel("Train Participant")
plt.tight_layout()
plt.savefig("decoder_transfer_matrix_PERM1_sameScaleAsTRUE.png", dpi=200)
plt.show()

# -----------------------------
# Distributions of pair scores (off-diagonal)
# -----------------------------
true_pairs = offdiag_values(score_matrix_true)
perm_pairs = offdiag_values(score_matrix_perm)

plt.figure(figsize=(8, 5))
bins = np.linspace(min(true_pairs.min(), perm_pairs.min()),
                   max(true_pairs.max(), perm_pairs.max()), 30)
plt.hist(true_pairs, bins=bins, alpha=0.6, label="TRUE", density=True)
plt.hist(perm_pairs, bins=bins, alpha=0.6, label="PERM (1)", density=True)
plt.axvline(90, color='red', linestyle='--', label='Chance (90)')
plt.xlabel("Mean Rank (lower = better)")
plt.ylabel("Density")
plt.title("Distribution of off-diagonal pair scores\nTRUE vs Permuted labels (1 perm)")
plt.legend()
plt.tight_layout()
plt.savefig("pair_scores_distributions_TRUE_vs_PERM1.png", dpi=200)
plt.show()

# -----------------------------
# Save matrices
# -----------------------------
score_df_true_with_avg.to_csv("decoder_transfer_matrix_TRUE_with_avg.csv")
score_df_perm_with_avg.to_csv("decoder_transfer_matrix_PERM1_with_avg.csv")
print("✅ Saved: TRUE & PERM heatmaps (same scale), distributions, and CSVs.")
```

```
import numpy as np
import pandas as pd
from scipy.stats import f

# -------------- helpers --------------
def offdiag_values(mat: np.ndarray) -> np.ndarray:
    """Return flattened off-diagonal values from a square matrix."""
    n = mat.shape[0]
    mask = ~np.eye(n, dtype=bool)
    return mat[mask]

# -------------- build TRUE matrix --------------
n = len(participant_ids)
score_matrix_true = np.zeros((n, n), dtype=float)

print("\n=== Computing TRUE score matrix ===")
for i in range(n):
    for j in range(n):
        score = crossval_17_1_decoder_eval(
            X_train=participant_data[i],
            Y_train=vectors,
            X_test=participant_data[j],
            Y_test=vectors
        )
        score_matrix_true[i, j] = score
        print(f"[TRUE] Decoder {participant_ids[i]} → Test {participant_ids[j]}: mean rank = {score:.2f}")

true_pairs = offdiag_values(score_matrix_true)

# -------------- F-tests over 50 permutations --------------
rng = np.random.default_rng(1234)
n_perms = 50
alpha = 0.05
n_different = 0

print("\n=== Variance tests: TRUE vs single-permutation null (50 runs) ===")

for p in range(1, n_perms + 1):
    perm_idx = rng.permutation(vectors.shape[0])
    perm_vectors = vectors[perm_idx]

    score_matrix_perm = np.zeros((n, n), dtype=float)
    for i in range(n):
        for j in range(n):
            score = crossval_17_1_decoder_eval(
                X_train=participant_data[i],
                Y_train=perm_vectors,
                X_test=participant_data[j],
                Y_test=vectors
            )
            score_matrix_perm[i, j] = score

    perm_pairs = offdiag_values(score_matrix_perm)

    # --- F-test for equality of variances ---
    var_true = np.var(true_pairs, ddof=1)
    var_perm = np.var(perm_pairs, ddof=1)

    if var_true > var_perm:
        F_stat = var_true / var_perm
        p_val = 1 - f.cdf(F_stat, len(true_pairs)-1, len(perm_pairs)-1)
    else:
        F_stat = var_perm / var_true
        p_val = 1 - f.cdf(F_stat, len(perm_pairs)-1, len(true_pairs)-1)

    is_diff = p_val < alpha
    n_different += int(is_diff)

    print(f"Perm {p:02d}: F = {F_stat:.4f}, p = {p_val:.4g} --> "
          f"{'DIFFERENT' if is_diff else 'not different'} at α={alpha}")

# -------------- summary --------------
percent_diff = 100.0 * n_different / n_perms
print(f"\n=== Summary ===")
print(f"Different variances in {n_different}/{n_perms} tests "
      f"({percent_diff:.1f}%).")
```

```
import numpy as np
import pandas as pd
import glob
import os
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import random

# -----------------------------
# Load concept labels (expecting 180)
# -----------------------------
concepts = np.genfromtxt("stimuli_180concepts.txt", dtype=str)
assert len(concepts) == 180, "Expected 180 concepts."

# -----------------------------
# Load participant fMRI data
# -----------------------------
file_paths = sorted(glob.glob("M*_avg_pca.csv"))
participant_ids = [os.path.basename(f).split('_')[0] for f in file_paths]
participant_data = {pid: pd.read_csv(f).values for pid, f in zip(participant_ids, file_paths)}

# -----------------------------
# Pick 2 concepts at random to highlight
# -----------------------------
selected_concepts = random.sample(list(concepts), 2)

# -----------------------------
# Collect PCA-projected coordinates for the two selected concepts across all participants
# -----------------------------
highlight_df = []

for pid in participant_ids:
    X = participant_data[pid]
    if X.shape[0] != len(concepts):
        print(f"Skipping {pid}: unexpected shape {X.shape}")
        continue

    # Normalize and apply PCA
    X_norm = StandardScaler().fit_transform(X)
    X_pca = PCA(n_components=2).fit_transform(X_norm)

    df = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
    df["concept"] = concepts
    df = df[df["concept"].isin(selected_concepts)]
    df["participant"] = pid
    highlight_df.append(df)

# -----------------------------
# Combine and plot
# -----------------------------
if highlight_df:
    plot_df = pd.concat(highlight_df, ignore_index=True)

    plt.figure(figsize=(8, 6))
    sns.scatterplot(data=plot_df, x="PC1", y="PC2", hue="concept", style="participant", s=80)
    plt.title(f"2D PCA Plot for Concepts: {selected_concepts[0]} and {selected_concepts[1]}")
    plt.legend(title="Concept / Participant", bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.tight_layout()
    plt.show()
else:
    print("No valid participant data matched the selected concepts.")

```
